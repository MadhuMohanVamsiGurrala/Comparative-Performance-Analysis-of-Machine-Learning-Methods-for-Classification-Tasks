```python
import numpy as np
import matplotlib.pyplot as plt

# Data
methods = ['XGBoost', 'LightGBM', 'CatBoost', 'MLP', 'LR', 'ID3']
accuracy = [87.10, 85.10, 88.70, 90.50, 72.50, 93.90]
precision = [88.90, 84.00, 88.70, 86.70, 59.20, 90.30]
recall = [83.40, 81.10, 82.60, 78.50, 50.70, 90.00]
f1_score = [85.80, 82.50, 85.20, 81.30, 52.60, 90.20]
auc_roc = [98.70, 97.30, 98.90, 99.20, 93.60, 94.30]
auc_pr = [93.60, 85.50, 93.10, 90.50, 60.70, 82.40]
log_loss = [32.30, 51.10, 30.00, 23.30, 63.20, 20.30]
kappa = [79.20, 76.00, 81.70, 84.80, 54.80, 90.20]
brier_score = [2.70, 3.10, 2.50, 1.90, 5.50, 1.70]
mcc = [79.20, 76.10, 81.70, 84.80, 54.90, 90.20]
fmi = [77.50, 74.80, 80.50, 84.40, 61.10, 89.40]
r2 = [76.10, 68.00, 74.40, 74.80, 22.70, 84.80]
top_k_acc = [100.00, 99.90, 100.00, 99.90, 99.40, 94.40]

metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC-ROC', 'AUC-PR', 'Log Loss', "Cohen's Kappa", 'Avg Brier Score', 'MCC', 'Fowlkes-Mallows Index', 'R^2', 'Top-K Accuracy']
values = [accuracy, precision, recall, f1_score, auc_roc, auc_pr, log_loss, kappa, brier_score, mcc, fmi, r2, top_k_acc]

# Convert data to polar coordinates
angles = np.linspace(0, 2 * np.pi, len(metrics), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))  # Close the circle

# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# Plot each method
for i, method in enumerate(methods):
    values_polar = np.concatenate((np.array(values)[:, i], [np.array(values)[:, i][0]]))
    values_polar = 2 * np.pi * (values_polar - np.min(values_polar)) / (np.max(values_polar) - np.min(values_polar))
    ax.plot(angles, values_polar, label=method)
    ax.fill(angles, values_polar, alpha=0.1)

# Set labels and title
ax.set_xticks(angles[:-1])
ax.set_xticklabels(metrics)
ax.set_title('Comparison of Methods across Different Metrics', pad=20)

# Add legend
ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```
